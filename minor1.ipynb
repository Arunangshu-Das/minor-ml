```python
from sklearn.impute import KNNImputer
imputer=KNNImputer(n_neighbors=20)
df=imputer.fit_transform(df.drop(['Disease'],axis=1))
```

```python
df
```

```python
df=pd.DataFrame(df)
df.columns=diseases.columns[1:]
```

```python
df.info()
```

```python
for i in df.columns:
    df[i] = df[i].astype(np.int64)
```

```python
df['Diseases']=diseases['Disease']
```

```python
df
```

```python
# !pip install yellowbrick
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop(['Diseases'],axis=1),df['Diseases'],test_size=0.2,random_state=1234)
```

```python
from yellowbrick.cluster import KElbowVisualizer
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans
```

```python
model=KMeans()
visualizer = KElbowVisualizer(model, k=(4,50))
```

```python
visualizer.fit(x_train,y_train)        # Fit the data to the visualizer
visualizer.show() 
```

```python
model=KMeans(n_clusters=41)
model.fit(x_train)
```

```python
labels = model.predict(x_test)  
```

```python
centroids = model.cluster_centers_ 
```

```python
labels=model.predict(x_train)
```

```python
labels
```

```python
k=[]
for i in np.unique(labels):
    k.append(np.where(labels==i)[0][0])
```

```python
ans=y_train.iloc[k].values
```

```python
cluster_name=pd.DataFrame()
cluster_name['idx']=np.unique(labels)
cluster_name['Diseases']=ans
```

```python
cluster_name
```

```python
y_pred_num=model.predict(x_test)
y_pred=[]
for i in y_pred_num:
    y_pred.append(cluster_name['Diseases'][cluster_name['idx']==i].values[0])
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
